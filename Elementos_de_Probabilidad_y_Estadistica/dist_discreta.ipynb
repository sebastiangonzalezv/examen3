```r
#-----------------------------------------------------------------
# The url with data is:
enlace <- "http://www.stat.berkeley.edu/~statlabs/data/crabs.data"
# R can open the url with the next code:
#browseURL(enlace)

# Loading the data in R:
datos <- read.table(enlace, header=T)
# To see the head of the data
head(datos)
# To obtain the density plot
plot(density(datos$presz[datos$year==81],na.rm=T), col='blue', lwd=2,
ylim=c(0, 0.04), main='Density plot for size by year',
xlab='Size of the carapace (mm)')
lines(density(datos$presz[datos$year==82],na.rm=T), col='red', lwd=2)
lines(density(datos$presz[datos$year==92],na.rm=T), col='orange', lwd=2)
legend('topleft', legend=c('81','82','92'), col=c('blue','red','orange'), lwd=2)
# To obtain the empirical CDF
cdf81 <- ecdf(datos$presz[datos$year==81])
cdf82 <- ecdf(datos$presz[datos$year==82])
cdf92 <- ecdf(datos$presz[datos$year==92])
plot(cdf81, col='blue', main='Empirical cdf for size by year',
xlab='Size of the carapace (mm)')
lines(cdf82, col='red')
lines(cdf92, col='orange')
legend('bottomright', legend=c('81','82','92'), col=c('blue','red','orange'), lwd=2)
```
