### Examen de la resistencia de pañuelos
Las grandes empresas solicitan inconformidades a sus consumidores con el fin de identificar y tratar decorregir los problemas en el proceso de fabricación que producen las inconformidades. Una de las quejasmás comunes ante un gran fabricante de pañuelos faciales es que las hojas se rompen al retirarlas de lacaja. Como la resistencia a la tracción es el factor que determina cuándo se rasgará una hoja, la gerenciadecidió re-examinar una variable muy importante del proceso que afecta la resistencia a la tracción, que esla resistencia en la dirección transversal (RDT). La RDT es la fuerza en la dirección en la que el pañuelo sesaca de la caja. Un valor de la RDT que sea demasiado alto crea una hoja rígida, mientras que un valor dela RDT que sea demasiado bajo puede hacer que las hojas se rasguen al retirarlas de la caja.

Según las especificaciones, se supone que la RDT sigue una distribución normal con una media de 300lb/resma y una desviación estándar de 4 lb/resma. Los valores críticos para la RDT son 290 y 315. Unahoja con un valor de RDT inferior a 290 lb/resma o superior a 315 lb/resma se considera inaceptable. (unaresma está formada por 480 hojas individuales o pañuelos faciales)

```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Suponga que $\mu = 300$ y $\sigma = 4$. ¿Qué fracción de pañuelos tendrá una RDT inaceptable(RDT por debajo de 290 o por encima de 315 lb/resma)?.  ¿Qué fracción tendría una RDTinaceptable si el valor de $\mu$ fuera de 298 con el mismo valor de $\sigma$?.

```python
probability_300_4 = 1 - st.norm(300, 4).cdf(315) + st.norm(300, 4).cdf(290)
probability_298_4 =  1 - st.norm(298, 4).cdf(315) + st.norm(298, 4).cdf(290)
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

x_1 = np.arange(276, 320, 0.001)
y_1 = st.norm(300, 4).pdf(x_1)
ax1.plot(x_1, y_1)

x_low = np.arange(276, 290, 0.001)
y_low = st.norm(300, 4).pdf(x_low)
ax1.fill_between(x_low, y_low, color='r')

x_up = np.arange(315, 320, 0.001)
y_up = st.norm(300, 4).pdf(x_up)
ax1.fill_between(x_up, y_up, color='r')

x_1 = np.arange(276, 320, 0.001)
y_1 = st.norm(298, 4).pdf(x_1)
ax2.plot(x_1, y_1)

x_low = np.arange(276, 290, 0.001)
y_low = st.norm(298, 4).pdf(x_low)
ax2.fill_between(x_low, y_low, color='r')

x_up = np.arange(315, 320, 0.001)
y_up = st.norm(298, 4).pdf(x_up)
ax2.fill_between(x_up, y_up, color='r')

print(probability_300_4)
print(probability_298_4)
plt.show()
```

si: $\mu = 300$ y $\sigma = 4$ la probabilidad de tener un RTD inaceptable es de 0.63%<br> 
si: $\mu = 300$ y $\sigma = 6$ la probabilidad de tener un RTD inaceptable es de 5.40%<br> 

(2pts) Ahora suponga que la RDT media es 300 con el valor de $\sigma$ igual a 6. ¿Qué fracción depañuelos será inaceptable?.

```python
probability_300_6 =  1 - st.norm(300, 6).cdf(315) + st.norm(300, 6).cdf(290)

fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(111)

x_1 = np.arange(276, 320, 0.001)
y_1 = st.norm(300, 6).pdf(x_1)
ax1.plot(x_1, y_1)

x_low = np.arange(276, 290, 0.001)
y_low = st.norm(300, 6).pdf(x_low)
ax1.fill_between(x_low, y_low, color='r')

x_up = np.arange(315, 320, 0.001)
y_up = st.norm(300, 6).pdf(x_up)
ax1.fill_between(x_up, y_up, color='r')

print(probability_300_6)
plt.show()
```

si: $\mu = 300$ y $\sigma = 6$ la probabilidad de tener un RTD inaceptable es de 5.40%<br> 

(3pts) ¿Cuál es la fracción de pañuelos con el valor de la RDT dentro de 1 desviación estándar dela media de 300? ¿Cuál es la fracción de pañuelos dentro de 2 desviaciones estándar de la media? Suponga acá que $\sigma= 4$ lb/resma.

```python
probability_300_4_1_sigma = st.norm(300, 4).cdf(300+4) - st.norm(300, 4).cdf(300-4)
probability_300_4_2_sigma = st.norm(300, 4).cdf(300+8) - st.norm(300, 4).cdf(300-8)
print(probability_300_4_1_sigma)
print(probability_300_4_2_sigma)

fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

x_1 = np.arange(276, 320, 0.001)
y_1 = st.norm(300, 4).pdf(x_1)
ax1.plot(x_1, y_1)

x_mid = np.arange(296, 304, 0.001)
y_mid = st.norm(300, 4).pdf(x_mid)
ax1.fill_between(x_mid, y_mid, color='r')

ax2.plot(x_1, y_1)

x_mid = np.arange(292, 308, 0.001)
y_mid = st.norm(300, 4).pdf(x_mid)
ax2.fill_between(x_mid, y_mid, color='r')

plt.show()
```

(3pts) Suponga $\mu= 300 $ y $\sigma= 4$. ¿Cuál es el valor de la RDT que es superado por el 95% de los pañuelos? ¿Cuál sería el valor de la RDT que es superado por el 99% de los pañuelos?

```python
probability_300_4_95 = st.norm(300, 4).interval(0.95)
probability_300_4_99 = st.norm(300, 4).interval(0.99)
print(probability_300_4_95)
print(probability_300_4_99)
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

x_1 = np.arange(276, 320, 0.001)
y_1 = st.norm(300, 4).pdf(x_1)
ax1.plot(x_1, y_1)

x_mid = np.arange(probability_300_4_95[0], probability_300_4_95[1], 0.001)
y_mid = st.norm(300, 4).pdf(x_mid)
ax1.fill_between(x_mid, y_mid, color='r')

ax2.plot(x_1, y_1)

x_mid = np.arange(probability_300_4_99[0], probability_300_4_99[1], 0.001)
y_mid = st.norm(300, 4).pdf(x_mid)
ax2.fill_between(x_mid, y_mid, color='r')

plt.show()
```

```python
np.random.seed(1037626439)
random_sample = np.random.normal(loc=300, scale=4, size=200)
good_quality = np.extract(random_sample < 315, random_sample)
good_quality = np.extract(good_quality > 290, good_quality)
probabilty_bad_quality = 1- (len(good_quality)/len(random_sample))
standar_desviation = np.sqrt(np.var(random_sample, ddof=1))

good_quality_1_std = np.extract(random_sample < random_sample.mean()+standar_desviation, random_sample)
good_quality_1_std = np.extract(good_quality_1_std > random_sample.mean()-standar_desviation, good_quality_1_std)
probabilty_good_quality_1_std = (len(good_quality_1_std)/len(random_sample))

good_quality_2_std = np.extract(random_sample < random_sample.mean()+2*standar_desviation, random_sample)
good_quality_2_std = np.extract(good_quality_2_std > random_sample.mean()-2*standar_desviation, good_quality_2_std)
probabilty_good_quality_2_std = (len(good_quality_2_std)/len(random_sample))

good_quality_3_std = np.extract(random_sample < random_sample.mean()+3*standar_desviation, random_sample)
good_quality_3_std = np.extract(good_quality_3_std > random_sample.mean()-3*standar_desviation, good_quality_3_std)
probabilty_good_quality_3_std = (len(good_quality_3_std)/len(random_sample))

z = (random_sample-random_sample.mean())/standar_desviation
z_mean = z.mean()
z_std = np.sqrt(np.var(z, ddof=1))
print('{}\n{}\n{}\n{}\n{}\n{}'.format(probabilty_bad_quality, probabilty_good_quality_1_std, probabilty_good_quality_2_std, probabilty_good_quality_3_std, z_mean, z_std))
```
